```python
password = input("Enter the pass: ")
digits = "0123456789"
spec_char = "@#$%&*"
has_special = False
for ch in password:
    if 'A'<=ch<='Z':  
        continue
    elif 'a'<=ch<='z':  
        continue
    elif ch in digits:      
        continue
    elif ch in spec_char:  
        has_special=True
        continue
    else:
        print("Invalid password")
        break
else:
    if has_special:
        print("valid")
    else:
        print("Invalid password")

```
