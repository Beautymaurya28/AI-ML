```python
#Q2. accept the string from the user and display : the number of vowels and consonants ,the reversed string , whether it is as palindrome or not
#isalpha
str1=input("enter the string: ")
#count vowel and consonants
vowel="aeiou"
vow_count=0
cons_count=0
for ch in str1:
    if ch in vowel:
        vow_count+=1
    else:
        cons_count+=0
print(vow_count,cons_count)

#reversed the string
rev_str=str1[::-1]
print(rev_str)

#palindrome
if str1==rev_str:
    print("yes plaindrome")
else:
    print("No, this string is not plaindrome.")



```
