```python
#5 take a sentence as a input and count the frequency of each word using a dictionary 
sent=input("enter the sentence: ")
words=sent.split()
print(words)
freq={}
for word in words:
    if word in freq:
        freq[word]+=1
    else:
        freq[word]=1
for key,values in freq.items():
    print(f"{key}:{values}")



```

```python
#7 given two sets A and B .find union,interaction,difference and symmetric
setA={1,3,5,6,8,9,3}
setB={5,"apple","a",59,984}
unionset=setA.union(setB)
print(unionset)
intersection_set=setA.intersection(setB)
print(intersection_set)
diff_set=setA.difference(setB)
print(diff_set)
symm_set=setA.symmetric_difference(setB)
print(symm_set)
```
