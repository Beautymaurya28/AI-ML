```python
#write a program to short the list of string in alphabetical order without using short method\
my_list=["name","apple","orange","grapes"]
n=len(my_list)
for i in range(n):
    for j in range(0,n-i-1):
        if(my_list[j]>my_list[j+1]):
            my_list[j],my_list[j+1]=my_list[j+1],my_list[j]
print(my_list)
```

```python
fruits = ["apple", "banana", "apple", "mango", "banana", "apple"]
count_dict={}
for fruit in fruits:
    if fruit in count_dict:
        count_dict[fruit]+=1
    else:
        count_dict[fruit]=1
print(count_dict)
```
